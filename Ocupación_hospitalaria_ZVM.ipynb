# Ocupación Hospitalaria de la Zona del Valle de México

```python
###### ------------------- Main libraries -------------------################
import pandas as pd
import numpy as np 
from pprint import pprint
import folium 
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import datetime
```

```python
hospitales = pd.read_csv('./ocupación_hospitalaria.csv')
hospitales.head()
```

```python
hospitales.Estatus_capacidad_hospitalaria.value_counts()
```

```python
hospitales['color'] = np.where(hospitales.Estatus_capacidad_hospitalaria=='Buena','green',
                               np.where(hospitales.Estatus_capacidad_hospitalaria=='CrÃ­tica','red',
                                       np.where(hospitales.Estatus_capacidad_hospitalaria=='Media','blue','darkblue')))
hospitales.head(2)
```

```python
map1 = folium.Map(location = [19.4978,-99.1269], zoom_start = 10)
locations = hospitales[['lat', 'long']]
locationlist = locations.values.tolist()
#for point in range(0,len(locationlist)):
#    folium.Marker(locationlist[point], popup=[general_ags['MUNICIPIO'],general_ags.DEFUNCION]).add_to(map1)
#map1
marker_cluster = MarkerCluster().add_to(map1)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=hospitales['Estatus_capacidad_hospitalaria'][point],
                 icon=folium.Icon(color=hospitales["color"][point], icon_color='white', icon='male', angle=0, prefix='fa')).add_to(marker_cluster)
#map1.save('./mx_covid.html')
map1
```
